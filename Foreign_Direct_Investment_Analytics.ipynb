```python
# Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('/content/FDI data.csv')
```

```python
df
```

```python
# Data Cleaning and Preprocessing
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
print(df.dtypes)
```

```python
df.columns
```

```python
# Exploratory Data Analysis (EDA)
#Sector-wise investment analysis
# Melt the dataframe to have 'Year' and 'Amount' columns
df_melted = df.melt(id_vars=["Sector"], var_name="Year", value_name="Amount")

df_melted
```

```python
# Group by sector and sum the investment amounts
sector_investment = df_melted.groupby('Sector')['Amount'].sum().reset_index()

# Display the result
print(sector_investment)
```

```python
# Year-wise investment analysis
# Group by year and sum the investment amounts
year_investment = df_melted.groupby('Year')['Amount'].sum().reset_index()

# Display the result
print(year_investment)
```

```python
# Visualization
# Sector-wise investment

# Sort the data by investment amount
sector_investment_sorted = sector_investment.sort_values(by='Amount', ascending=False)

# Create a color palette
palette = sns.color_palette("viridis", len(sector_investment_sorted))

# Create the bar plot
plt.figure(figsize=(14, 10))
barplot = sns.barplot(
    x='Amount',
    y='Sector',
    data=sector_investment_sorted,
    palette=palette
)

# Add data labels
for i in barplot.containers:
    barplot.bar_label(i, fmt='%.2f', padding=3)

# Set title and labels
plt.title('Sector-wise FDI in India', fontsize=16)
plt.xlabel('Investment Amount (in million USD)', fontsize=14)
plt.ylabel('Sector', fontsize=14)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Year-wise investment
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Amount', data=year_investment)
plt.title('Year-wise FDI in India')
plt.xlabel('Year')
plt.ylabel('Investment Amount')
plt.xticks(rotation=45)
plt.show()
```
